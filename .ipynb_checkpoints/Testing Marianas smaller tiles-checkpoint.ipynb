# Investigate Marianas seeding issue
The original dataset had problem with unstable training. One way to combat this is to increase the batch size. However the tiles with the size 500x500 prevented batch sizes large than 6 due to memory limits on the GPU. To investigate this further I split the data into tiles of the size 250x250 and trained models with increased batch size.

The first step is to split the tiles into chips of the size 250x250 pixels

```python
# split hpmf
!python /workspace/code/tools/split_training_data.py /workspace/data/hpmf/ /workspace/data/hpmf250/ --tile_size 250

# Split labels
!python /workspace/code/tools/split_training_data.py /workspace/data/final_output/ /workspace/data/labels250/ --tile_size 250
```

## Remove unlabeled chips
Mabye of the splitted chips do not contain any pixels labeled as ditches. In an attempt to balance the dataset somewhat we deided to remove chips that contain less than 62 labeled ditch pixels.

```python
!python /workspace/code/tools/remove_unlabled_chips.py 62 /workspace/data/labels250/ /workspace/data/hpmf250/
```

I plotted the image chips to make sure that the splitting process worked as intended. The first and last chips are inspected.

```python
import glob
import matplotlib.pyplot as plt
import tifffile
plt.rcParams['figure.figsize'] = [16, 16]

split_labels = glob.glob('/workspace/data/labels250/*.tif')
split_hpmf = glob.glob('/workspace/data/hpmf250/*.tif')

split_labels.sort
split_hpmf.sort

print('The first label chip is named', split_labels[0])
print('The first hpmf chip is named', split_hpmf[0])
print('The last label chip is named', split_labels[-1])
print('The last hpmf chip is named', split_hpmf[-1])


f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(tifffile.imread(split_labels[0]))
axarr[1,0].imshow(tifffile.imread(split_labels[-1]))

axarr[0,1].imshow(tifffile.imread(split_hpmf[0]))
axarr[1,1].imshow(tifffile.imread(split_hpmf[-1]))
```
