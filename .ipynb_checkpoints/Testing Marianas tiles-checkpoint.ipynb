# Investigation of Marianas data

Mariana was having some issues with splitting tiles and getting inconsistent results. She have noted that running the split script multiple times result in different results of the Unet model. This notebook will try to replicate that behaviour. I will split the data twice and then train two unet models to see if there is any difference.

Fist I made sure that both labels and images contains the same number of tiles and that the names match

```python
labels = glob.glob('/workspace/data/final_output/*.tif')
print(len(labels), ' labeled tiles')
hpmf = glob.glob('/workspace/data/hpmf/*.tif')
print(len(hpmf), ' hpmf image tiles')
```

```python
import glob
import matplotlib.pyplot as plt
import tifffile


plt.rcParams['figure.figsize'] = [16, 16]
labels.sort
hpmf.sort
print('The first label tile is named', labels[0])
print('The first hpmf tile is named', hpmf[0])

f, axarr = plt.subplots(2,2)
axarr[0,0].imshow(tifffile.imread(labels[0]))
axarr[1,0].imshow(tifffile.imread(labels[10]))

axarr[0,1].imshow(tifffile.imread(hpmf[10]))
axarr[1,1].imshow(tifffile.imread(hpmf[0]))
```

## Seed 1

```python
# size of plot
plt.rcParams['figure.figsize'] = [16, 16]
f, axarr = plt.subplots(1,3)

#image 
orgimage = tifffile.imread('/workspace/data/inference_unet/test_chip/0001.tif')
# label
testlabel = tifffile.imread('/workspace/data/split_labels/0001.tif')
# prediction
prediction = tifffile.imread('/workspace/data/inference_unet/model1/0001.tif')
# Image
axarr[0].set_title('Original image')
axarr[0].imshow(orgimage, cmap='Greys_r')
# Label
axarr[1].set_title('Labeled Ditches')
axarr[1].imshow(testlabel)
# Prediction
axarr[2].set_title('Predicted Ditches')
axarr[2].imshow(prediction>0.5)
```

```python
plt.rcParams['figure.figsize'] = [16, 16]
f, axarr = plt.subplots(1,2)

axarr[0].set_title('Original image and labels')
axarr[0].imshow(orgimage, cmap='Greys_r')
axarr[0].imshow(testlabel, alpha=0.7)

axarr[1].set_title('Original image and predictions')
axarr[1].imshow(orgimage, cmap='Greys_r')
axarr[1].imshow((prediction>0.5), alpha=0.7)
```
