```python
import pandas as pd
import numpy as np
eval = pd.read_csv('/workspace/data/logfiles/test/basic_unet/test.csv')
TP = np.longdouble(eval['1_tp'].sum())
FP = np.longdouble(eval['1_fp'].sum())
TN = np.longdouble(eval['1_tn'].sum())
FN = np.longdouble(eval['1_fn'].sum())

MCC = ((TP*TN) - (FP*FN)) / np.sqrt((TP+FP) * (TP+FN) * (TN+FN) * (TN+FN))
MCC
```
