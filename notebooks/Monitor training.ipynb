```python
%%capture
import os
import seaborn as sns
import numpy as np
import pandas as pd
logdir = '//workspace/data/logfiles/UNet/05m/'

data1 = pd.DataFrame(
    {
        "index": [],
        "Precision": [],
        "Recall": [],
        "MCC": [],
        "F1": [],
        "IoU": [],
        "TP": [],
        "FP": [],
        "TN": [],
        "FN": [],
    }
)


for subdir, dirs, csvs in os.walk(logdir):
    for file in csvs:
        if file.endswith('test.csv'):
            result = pd.read_csv(subdir + '/' + file)
            index = os.path.split(subdir)[1]
            TP = np.longdouble(result['1_tp'].sum())
            FP = np.longdouble(result['1_fp'].sum())
            TN = np.longdouble(result['1_tn'].sum())
            FN = np.longdouble(result['1_fn'].sum())
            
            MCC = ((TP*TN) - (FP*FN)) / np.sqrt((TP+FP) * (TP+FN) * (TN+FN) * (TN+FN))
            Precision = (TP / (TP + FP))
            Recall = (TP / (TP + FN))
            F1 = 2*(TP / (TP + FP))*(TP / (TP + FN))/((TP / (TP + FP))+(TP / (TP + FN)))
            IoU = TP / (TP + FP + FN)
            row1 = pd.Series([index,Precision, Recall, MCC, F1, IoU, TP, FP, TN, FN], index=data1.columns)
            data1 = data1.append(row1,ignore_index=True)
            #data1 = pd.concat([data1,row1])
            print(index)
            print(MCC)



data1.to_csv('/workspace/data/logfiles/UNet/05m/05m.csv', index=False) 
```

```python
import pandas as pd
import numpy as np
eval = pd.read_csv('/workspace/data/logfiles/UNet/1m/hillshade1/test.csv')
TP = np.longdouble(eval['1_tp'].sum())
FP = np.longdouble(eval['1_fp'].sum())
TN = np.longdouble(eval['1_tn'].sum())
FN = np.longdouble(eval['1_fn'].sum())

MCC = ((TP*TN) - (FP*FN)) / np.sqrt((TP+FP) * (TP+FN) * (TN+FN) * (TN+FN))
MCC
```
