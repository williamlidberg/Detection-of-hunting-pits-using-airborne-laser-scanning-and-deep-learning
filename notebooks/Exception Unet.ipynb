## Collect results from logs

```python
%%capture
import os
import seaborn as sns
import numpy as np
import pandas as pd
logdir = '/workspace/data/logfiles/05m/'

data1 = pd.DataFrame(
    {
        "index": [],
        "MCC": [],
        "TP": [],
        "FP": [],
        "TN": [],
        "FN": [],
    }
)


for subdir, dirs, csvs in os.walk(logdir):
    for file in csvs:
        if file.endswith('test.csv'):
            result = pd.read_csv(subdir + '/' + file)
            index = os.path.split(subdir)[1]
            TP = np.longdouble(result['1_tp'].sum())
            FP = np.longdouble(result['1_fp'].sum())
            TN = np.longdouble(result['1_tn'].sum())
            FN = np.longdouble(result['1_fn'].sum())
            
            MCC = ((TP*TN) - (FP*FN)) / np.sqrt((TP+FP) * (TP+FN) * (TN+FN) * (TN+FN))
            
            row1 = pd.Series([index, MCC, TP, FP, TN, FN], index=data1.columns)
            data1 = data1.append(row1,ignore_index=True)
            #data1 = pd.concat([data1,row1])
            print(index)
            print(MCC)


data1.to_csv('/workspace/data/logfiles/05m.csv', index=False)          

```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

results_05m = pd.read_csv('/workspace/data/logfiles/05m.csv', sep=',') # have a look at how the data is structured
results_1m = pd.read_csv('/workspace/data/logfiles/1m.csv', sep=',')

# add resolution column
results_05m['Resolution'] = '05m'
results_1m['Resolution'] = '1m'

MCC = results_05m.append(results_1m)

MCC['index'] = MCC['index'].str[:-1] # to remove seed number from directory name

sns.set_theme(style="ticks", palette="pastel")
sns.boxplot(data=MCC, y="index", x="MCC", hue="Resolution",orient="h")

plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel("MCC")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
plt.rcParams['figure.figsize'] = [8, 8]

MCC['Recall'] = MCC['TP'] / (MCC['TP'] + MCC['FN']) * 100
sns.set_theme(style="ticks", palette="pastel")
ax= sns.scatterplot(data=MCC, x="Recall", y="MCC", hue="index", size="Resolution",sizes={"05m": 20, "1m": 100})
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xticks(rotation=0)
plt.xticks(rotation=0)
plt.ylabel("MCC")
plt.xlabel("Recall")
plt.show()
```
