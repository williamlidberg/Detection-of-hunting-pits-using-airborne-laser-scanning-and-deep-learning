```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

def object_evaluation(label_dir, polygon_dir):
    TP_list = []
    FN_list = []
    FP_list = []
    
    for label in os.listdir(label_dir):
        if label.endswith('.shp'):
            if os.path.isfile(polygon_dir + label):
                gt = gpd.read_file(label_dir + label)
                pred = gpd.read_file(polygon_dir + label)
                gt.set_crs("EPSG:3006")
                pred.set_crs("EPSG:3006")
                for index, row in gt.iterrows():
                        TP = gt[gt.intersects(pred)] #intersect ground truth with predicted polygons.
                        if TP == 1:
                            TP_list.append(1)
                            FN_list.append(0)
                            FP_list.append(0)
                        if TP == 0:
                            TP_list.append(0)
                            FN_list.append(1)
                            FP_list.append(0)
            else:
                gt = gpd.read_file(label_dir + label)
                gt.set_crs("EPSG:3006")
                for index, row in gt.iterrows():
                    TP_list.append(0)
                    FN_list.append(1)
                    FP_list.append(0)
                    
                    
    for prediction in os.listdir(inference_polygon):
        if prediction.endswith('.shp'):
            gt = gpd.read_file(label_dir + prediction)
            pred = gpd.read_file(inference_polygon + prediction)
            gt.set_crs("EPSG:3006")
            pred.set_crs("EPSG:3006")

            for index, row in pred.iterrows():
                    polygon = pred[pred.intersects(gt)] #intersect ground truth with predicted polygons.
                    if predicted == 0:
                        FP_list.append(1)
                        TP_list.append(0)
                        FN_list.append(0)
  
    print(sum(TP_list))
    print(sum(FP_list))
    print(sum(FN_list))
    
label_dir = '/workspace/data/final_data_05m/testing/polygon_labels/'
inference_polygon = '/workspace/data/final_data_05m/testing/inference_XceptionUNet/05m/inference_polygon/minimal_curvature/'

object_evaluation(label_dir, inference_polygon)
```
