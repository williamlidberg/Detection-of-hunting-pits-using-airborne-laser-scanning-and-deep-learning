# Evaluate objects

```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

label_dir = '/workspace/data/final_data_05m/testing/polygon_labels/'
inference_polygon = '/workspace/data/final_data_05m/testing/inference_XceptionUNet/05m/inference_polygon/hillshade/'
results_list = []

for label in os.listdir(label_dir):
    if label.endswith('.shp'):
        try:
            gt = gpd.read_file(label_dir+label)
            pred = gpd.read_file(inference_polygon + label)
            #sindex = pred.sindex

            # iterate over the points
            for index, row in gt.iterrows():
                    # find approximate matches with r-tree, then precise matches from those approximate ones
                    TP = gt[gt.intersects(pred)]
                    results_list.append(len(TP))
                    print(label)
        except:
            print('file missing')
```
